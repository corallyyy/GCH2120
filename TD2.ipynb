```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from scipy.integrate import solve_ivp
```

```python
def bilans_composants(t, x, p, u):
    #Variable d'etat
    [TG, DG, MG, ME, GL, MeOH] = x
    #Parametres
    [k1, k2, k3, k4, k5, k6] = p
    #Eq. diff.
    dTG = -k1 * TG * MeOH + k2 * DG * ME
    dDG = k1 * TG * MeOH - k2 * DG * ME - k3 * DG * MeOH + k4 * MG * ME
    dMG = k3 * DG * MeOH - k4 * MG * ME - k5 * MG * MeOH + k6 * GL * ME
    dME = k1 * TG * MeOH - k2 * DG * ME + k3 * DG * MeOH - k4 * MG * ME + k5 * MG * MeOH - k6 * GL * ME
    dGL = k5 * MG * MeOH - k6 * GL * ME
    dMeOH = -k1 * TG * MeOH + k2 * DG * ME - k3 * DG * MeOH + k4 * MG * ME - k5 * MG * MeOH + k6 * GL * ME

    return [dTG, dDG, dMG, dME, dGL, dMeOH]
    
#Conditions initiales
x0 = [1.0, 0, 0, 0, 0, 3.0]
p = [0.12, 0.17, 0.61, 1.52, 2.56, 0.01]
t_sim = [0, 200]
u0 = []

sim = solve_ivp(bilans_composants, t_sim, x0, method = 'RK45', args = (p, u0), max_step = 0.01)

#Graphique
CTG, CDG, CMG, CME, CGL, CMeOH = sim.y

plt.figure(figsize=(8, 5), dpi=120)
plt.plot(sim.t, CTG, label = 'TG')
plt.plot(sim.t, CDG, label = 'DG')
plt.plot(sim.t, CMG, label = 'MG')
plt.plot(sim.t, CME, label = 'ME')
plt.plot(sim.t, CGL, label = 'GL')
plt.plot(sim.t, CMeOH, label = 'MeOH')
plt.xlabel('Temps [min]')
plt.ylabel('Concentration [mol/L]')
plt.title('Concentration de differents composants dans un procede de production de diesel')
plt.grid()
plt.legend()
plt.show()

#Valeur max de ME
max_ME = np.max(CME)
print(f"La concentration maximale du methyl-ester est de {max_ME:.4f} mol/L")

#Temps max
ME_95 = 0.95 * max_ME
indice_t= np.argmax(CME >= ME_95)
t_95 = sim.t[indice_t]
ME_95 = CME[indice_t]
print(f"On obtient cette concentration a {t_95:.4f} min")




```
